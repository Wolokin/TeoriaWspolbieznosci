```python
from collections import defaultdict, Counter
import matplotlib.pyplot as plt

def plot(dat, title):
    data = defaultdict(int)
    denom = Counter(dat[0])
    for s,t in zip(dat[0], dat[1]):
        data[s] += t/denom[s]

    lists = sorted(data.items())
    x, y = zip(*lists)
    print("total requests:", len(dat[0]))
    # print(sorted(y)[:10])
    plt.plot(x,y)
    plt.title(title)
    plt.ylabel("Average wait time [ns]")
    plt.xlabel("Block size")
    plt.show()

def plot_all(filename):
    with open(filename) as f:
        lines = f.readlines()
        for i in range(len(lines)):
            lines[i] = list(map(lambda x: int(x), lines[i][1:-2].replace(",","").split(" ")))
        producers = lines[:2]
        consumers = lines[2:]
    plot(producers, "Producers")
    plot(consumers, "Consumers")
```

```python
plot_all("naive_t1.out")
```

```python
plot_all("naive_t2.out")
```

```python
plot_all("naive_t3.out")
```
